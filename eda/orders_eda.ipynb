## Columns:

- order_id - unique identifier
- user_id - customer identifier
- eval_set - whether this is prior/train/test data
- order_number - which order this is for the customer (1st, 2nd, 3rd...)
- order_dow - day of week (0-6, where 0=Sunday)
- order_hour_of_day - hour (0-23)
- days_since_prior_order - days since last order (NaN for first orders)
